```python
# Authors: Group 18
# Nhan Kiet To (1043668)
# Adam Thalhammer (972267)
# Will Walters (583639)

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

sns.set(style="whitegrid")

plot_list = ["count-malignant-cells", "count-somatic-cells", "count-malignant-spawn-cells", "count-somatic-spawn-cells",
             "count-quiescent-cells-at-final-tick", "count-quiescent-cells-occurences", "count-vascular-patches"]
```

# Final state analysis

```python
df = pd.read_csv(
    'experiments/TumourAngioSim-final-state-table.csv',
    skiprows=6
)
for col_pl in plot_list:

    sns.violinplot(x="angiogenesis-signal-radius", y=f"{col_pl}", hue="initial-cells",
                   data=df, style="angiogenesis-min-demand")

    plt.savefig(f"draft-models/{col_pl}.png", dpi=96,
                transparent=True, format='png', bbox_inches = "tight")
    plt.show()
    plt.clf()
```

# Sensitivity Analysis

```python

df = pd.read_csv(
    'experiments/TumourAngioSim-final-state-table-2.csv',
    skiprows=6
)
for col_pl in plot_list:
    sns.violinplot(x="malignant-proliferation-chance", y=f"{col_pl}", hue="energy-req-prolif",
                   data=df, style="malignant-proliferation-chance")
    
    sns.violinplot(x="normal-proliferation-chance", y=f"{col_pl}", hue="energy-req-prolif",
                   data=df, style="malignant-proliferation-chance")


    plt.savefig(f"draft-models/sensitivity_{col_pl}.png", dpi=96, transparent=True, format='png', bbox_inches = "tight")
    plt.show()
    plt.clf()
```
